```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import json
from random import randrange
import time
import stanza
```

```python
stanza.download('tr')
nlp = stanza.Pipeline('tr')
```

```python
ex_sentence = "fıstık"
doc = nlp(ex_sentence)
for sent in doc.sentences:
  for word in sent.words:
    if word.upos == 'NOUN':
      print("evreka")

print(*[f'word: {word.text}\tupos: {word.upos}' for sent in doc.sentences for word in sent.words], sep='\n')
```

```python
json_list = []
start = time.time()
for i in range(600):
  while True:
    id = randrange(92400)
    flag = 0
    url_name = 'https://sozluk.gov.tr/gts_id?id={}'.format(id)
    headers = requests.utils.default_headers()

    headers.update(
        {
            'User-Agent': 'User Agent 1.0',
        }
    )
    data = requests.get(url_name, headers=headers)
    json_data = json.loads(data.content)
    root = json_data[0]['madde']
    #check space

    doc = nlp(root)
    for sent in doc.sentences:
      for word in sent.words:
        if word.upos != 'NOUN':
          flag = 1
          

    if ' ' in root or flag == 1:
      continue
    
    else:
      json_list.append(json_data[0]['madde'])
      break
print(json_list)
end = time.time()
print(end-start)
#content = json.loads(response.text)
#print(type(content))
```
